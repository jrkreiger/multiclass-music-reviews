# Introduction

This project examines [music reviews from Pitchfork (via Kaggle)](https://www.kaggle.com/nolanbconaway/pitchfork-data) to try to predict the genre of music reviewed based solely on the text of a review.

Here's an overview of my process:
1. Load and explore the data
2. Select genres to focus on (i.e., classes to predict)
3. Visualize relationships of chosen classes
4. Split data into train/validation/test sets
5. Vectorize the review texts using TFIDF
6. Balance classes in the training set
7. Use a baseline model to determine best number of features to use in TFIDF; repeat steps 5-7 as needed
8. Tune the baseline model
9. Try additional models

So far, the best model is getting the following F1 scores on the validation data:
- Class 0 (electronic): 0.80
- Class 1 (pop/r&b): 0.51
- Class 2 (rap): 0.80

This mix of F1 scores is not unexpected, since a TSNE visualization showed pretty clear separation of electronic and rap, with pop/r&b scattered all over.

## Load and explore the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# import chartify

import sqlite3

from collections import Counter
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (confusion_matrix, classification_report, auc, 
                             roc_curve, accuracy_score)
from sklearn.model_selection import GridSearchCV
```

```python
# Set up connection to the database
conn = sqlite3.connect('pitchfork.sqlite')
cur = conn.cursor()
```

```python
# Query the genres and reviews tables and load the data
cur.execute("""
    SELECT c.reviewid, c.content, g.genre
    FROM content AS c
    INNER JOIN genres AS g
    ON c.reviewid = g.reviewid
    WHERE g.genre IS NOT NULL
            ;""")
reviews = pd.DataFrame(cur.fetchall())
reviews.columns = [x[0] for x in cur.description]
reviews.head()
```

### Functions

```python
def print_metrics(y_true, y_pred):
    '''Print accuracy score, confusion matrix, and classification report.
    
    Keyword arguments:
    y_true: groud-truth labels
    y_pred: predicted labels
    
    Dependencies: 
    sklearn.metrics.accuracy_score
    sklearn.metrics.confusion_matrix
    sklearn.metrics.classification_report
    '''
    acc_score = accuracy_score(y_true, y_pred)
    conf_matrix = confusion_matrix(y_true, y_pred)
    class_report = classification_report(y_true, y_pred)
    print('Accuracy:', acc_score)
    print('-------')
    print('Confusion matrix:')
    print(conf_matrix)
    print('-------')
    print('Classification report:')
    print(class_report)
    
# Define a function to visualize feature importances
def plot_feature_importances(model, x):
    '''Display barplot representing a model's feature importances.
    
    Note that in a Jupyter Notebook, running this function will display
    the plot automatically.
    
    Keyword arguments:
    model: a trained classifier with .feature_importances_ attribute
    x: the training dataset or other object of identical dimensions
    
    Dependencies:
    matplotlib.pyplot aliased as plt
    '''
    n_features = x.shape[1]
    plt.figure(figsize=(8,10))
    plt.barh(range(n_features), model.feature_importances_, align='center') 
    plt.yticks(np.arange(n_features), range(1, n_features+1)) 
    plt.xlabel("Feature importance")
    plt.ylabel("Feature (Component)")
    plt.title('Feature Importances')
    
# Define a function to visualize a confusion matrix
def pretty_confusion(y_true, y_pred, model_name):
    '''Display normalized confusion matrix with color scale.
    
    Note that this function works specifically for data from the 
    Forest Cover Types dataset. 
    
    Keyword arguments:
    y_true: ground-truth labels
    y_pred: predicted labels
    model_name: name to print in the plot title
    
    Dependencies:
    numpy aliased as np
    sklearn.metrics.confusion_matrix
    matplotlib.pyplot aliased as plt
    seaborn aliased as sns
    '''
    
    # Calculate the confusion matrix
    matrix = confusion_matrix(y_true, y_pred)
    matrix = matrix.astype('float') / matrix.sum(axis=1)[:, np.newaxis]
    
    # Build the plot
    plt.figure(figsize=(16,7))
    sns.set(font_scale=1.4)
    sns.heatmap(matrix, annot=True, annot_kws={'size':10},
                cmap=plt.cm.Greens, linewidths=0.2)
    
    # Add labels to the plot
    class_names = ['electronic', 'pop/r&b', 'rap']
    tick_marks = np.arange(len(class_names))
    tick_marks2 = tick_marks + 0.5
    plt.xticks(tick_marks, class_names, rotation=25)
    plt.yticks(tick_marks2, class_names, rotation=0)
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title('Confusion Matrix for {}'.format(model_name)) 
    plt.tight_layout()
    
    
# Define a function to save a confusion matrix visualization    
def save_conf_matrix(y_true, y_pred, model_name):
    '''Save normalized confusion matrix with color scale as .png file.
    
    Note that in Jupyter Notebook, the plot will also be displayed
    automatically on calling this function. This function saves .png files
    with 300 dpi and 0.5-in. padding.
    
    Keyword arguments:
    y_true: ground-truth labels
    y_pred: predicted labels
    model_name: name to print in the plot title
    
    Dependencies:
    sklearn.metrics.confusion_matrix
    matplotlib.pyplot aliased as plt
    seaborn aliased as sns
    '''
    fig = pretty_confusion(y_true, y_pred, model_name)
    filename = '_'.join(model_name.split()) + '_confmatrix.png'
    plt.savefig(filename, pad_inches=0.5, dpi=300)

# Define function to plot ROC curves
def roc_it(y_true, y_pred, model_name, figsize=(12,10)):
    '''Plot ROC with a separate curve for each class.
    
        Keyword arguments:
    y_true: ground-truth labels
    y_pred: predicted labels
    model_name: name to print in the plot title
    
    Dependencies:
    pandas aliased as pd
    sklearn.metrics.roc_curve
    sklearn.metrics.auc
    matplotlib.pyplot aliased as plt
    '''
    y_true_dum = pd.get_dummies(y_true)
    y_pred_dum = pd.get_dummies(y_pred)
    n_classes = y_true.nunique()
    
    fprs = []
    tprs = []
    aucs = []

    for i in range(0,n_classes):
        fpr, tpr, _ = roc_curve(y_true_dum.loc[:,i], y_pred_dum.loc[:,i])
        fprs.append(fpr)
        tprs.append(tpr)
        aucs.append(round(auc(fpr, tpr),2))
        
    plt.figure(figsize=(12,10))
    for i in range(n_classes):
        plt.plot(fprs[i], tprs[i], label='Class: {}, AUC: {}'.format(i, 
                                                                     aucs[i]))
    plt.plot([0.0,1.0], [0.0,1.0], linestyle='--')
    plt.title('ROC curves by class for {}'.format(model_name))
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend();
```

## Select genres to focus on

```python
focused = reviews[reviews['genre'].isin(['electronic', 'rap', 'pop/r&b'])]
focused.info()
```

```python
focused.head()
```

```python
focused.genre.value_counts()
```

```python
# Convert genre to numeric? Yeah, that's how it was in the tree cover dataset
from sklearn.preprocessing import LabelEncoder

genres = list(focused.genre)

encoder = LabelEncoder()
labels = encoder.fit_transform(genres)

labels[:10]
```

```python
data = pd.concat([focused.reset_index(), pd.Series(labels, name='label')], axis=1)
data.drop('index', axis=1, inplace=True)
data.head()
```

```python
data.label.value_counts()
```

## Visualize relationships of chosen genres

```python
# Import needed packages
import yellowbrick
from sklearn.feature_extraction.text import TfidfVectorizer
from yellowbrick.text import TSNEVisualizer

# Prepare the data
tfidf = TfidfVectorizer(max_features=100)
X = tfidf.fit_transform(data.content)
y = data.label

# Plot t-SNE
tsne = TSNEVisualizer()
tsne.fit(X, y)
tsne.show();
```

## Split data into train/validation/test sets

```python
val_len = int(round(0.1 * len(data), 0))
```

```python
val_len
```

```python
# Train-val-test split
from sklearn.model_selection import train_test_split

y = data['label']
X = data['content']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=val_len,
                                                    stratify=y,
                                                    random_state=123)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, 
                                                  test_size=686,
                                                  stratify=y_train,
                                                  random_state=123)
```

## Vectorize with TFIDF

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string

stop_list = stopwords.words('english')
stop_list += list(string.punctuation)
stop_list += ['br', '.<', '..', '...', '``', "''", '--']
```

```python
# Now vectorize
from sklearn.feature_extraction.text import TfidfVectorizer

vecto = TfidfVectorizer(stop_words=stop_list, max_features=10000)
X_train_vec = vecto.fit_transform(X_train)

X_val_vec = vecto.transform(X_val)
X_test_vec = vecto.transform(X_test)
```

## Balance classes in training set

```python
y_train.value_counts()
```

```python
# Randomly undersample the larger classes
rus = RandomUnderSampler(random_state=3, 
                         sampling_strategy={0:1247})

X_rus, y_rus = rus.fit_resample(X_train_vec, y_train)

# Check class counts
Counter(y_rus)
```

```python
# Randomly oversample the smaller classes
smote = SMOTE(random_state=3, sampling_strategy={1:1247})

X_resampled, y_resampled = smote.fit_resample(X_rus, y_rus)

# Check class counts
Counter(y_resampled)
```

These are the datasets we have now:
- X_resampled, y_resampled: training set with classes balanced, vectorized, stopwords removed
- X_val_vec, y_val: validation set, vectorized, stopwords removed
- X_test_vec, y_test: test set, vectorized, stopwords removed

## Fit baseline model

```python
rf_1 = RandomForestClassifier(n_estimators=200, random_state=123)
rf_1.fit(X_resampled, y_resampled)

pred = rf_1.predict(X_val_vec)

print_metrics(y_val, pred)
```

```python
pred = rf_1.predict(X_resampled)
print_metrics(y_resampled, pred)
```

## Random Forest with PCA

```python
# Import TruncatedSVD (~ 4 min.)
from sklearn.decomposition import TruncatedSVD

# Try several numbers of components to explain 80% of variance
comps = [10, 50, 100, 200, 500, 1000, 2000, 5000]
var_sums = []
for comp in comps:
    svd = TruncatedSVD(n_components=comp, n_iter=10, random_state=42)
    svd.fit(X_resampled)
    var_sums.append(np.sum(svd.explained_variance_ratio_))
    
# Visualize results
plt.figure(figsize=(10,8))
plt.plot(comps, var_sums)
plt.xlabel('Number of components')
plt.ylabel('% of variance explained')
plt.title('Variance explained by n components')
plt.xticks([10, 50, 100, 200, 500, 1000, 2000, 5000])
plt.show();
```

```python
# Import TruncatedSVD (~ 4 min.)
from sklearn.decomposition import TruncatedSVD

# Try several numbers of components to explain 80% of variance
comps = [1700, 1800, 1900]
var_sums = []
for comp in comps:
    svd = TruncatedSVD(n_components=comp, n_iter=10, random_state=42)
    svd.fit(X_resampled)
    var_sums.append(np.sum(svd.explained_variance_ratio_))
    
# Visualize results
plt.figure(figsize=(10,8))
plt.plot(comps, var_sums)
plt.xlabel('Number of components')
plt.ylabel('% of variance explained')
plt.title('Variance explained by n components')
plt.xticks([1700, 1800, 1900])
plt.show();
```

```python
svd = TruncatedSVD(n_components=1700, n_iter=10, random_state=42)
X_train_svd = svd.fit_transform(X_resampled)

X_val_svd = svd.transform(X_val_vec)
X_test_svd = svd.transform(X_test_vec)
```

```python
rf_2 = RandomForestClassifier(n_estimators=100)
rf_2.fit(X_train_svd, y_resampled)

pred = rf_2.predict(X_train_svd)
print_metrics(y_resampled, pred)
```

```python
pred = rf_2.predict(X_val_svd)
print_metrics(y_val, pred)
```

```python
params = {'n_estimators': [100, 500, 1000],
          'max_depth': [5, 10, 25],
          'max_features': [100, 500, 1700]}

model = RandomForestClassifier()
grid = GridSearchCV(model, params, cv=3)

grid.fit(X_train_svd, y_resampled)

print('Best params:', grid.best_params_)
print('Best score:', grid.best_score_)
```
